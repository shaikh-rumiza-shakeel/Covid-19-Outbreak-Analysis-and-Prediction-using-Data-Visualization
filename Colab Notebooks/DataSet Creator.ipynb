```python
import pandas as pd
import os
import shutil
```

```python
import zipfile as zf
files = zf.ZipFile("covid-chestxray-dataset-master.zip", 'r')
files.extractall('COVID')
files.close()

files = zf.ZipFile("kaggle_chest_xray.zip", 'r')
files.extractall('COVID')
files.close()
```

```python
# Create data for positive samples (GitHub)

FILE_PATH = "/content/drive/MyDrive/covid-chestxray-dataset-master/covid-chestxray-dataset-master/metadata.csv"
IMAGES_PATH = "/content/drive/MyDrive/covid-chestxray-dataset-master/covid-chestxray-dataset-master/images"
```

```python
df = pd.read_csv(FILE_PATH)
print(df.shape)
```

```python
df.head()
```

```python
TARGET_DIR = "/content/drive/MyDrive/covid-chestxray-dataset-master/covid-chestxray-dataset-master/DataSet/Covid"

if not os.path.exists(TARGET_DIR):
    os.mkdir(TARGET_DIR)
    print("Covid folder created")
```

```python
cnt = 0

for (i,row)in df.iterrows():
    if row["finding"] == "Pneumonia/Viral/COVID-19" and row["view"]=="PA"and"AP":
       filename = row["filename"]
       image_path = os.path.join(IMAGES_PATH,filename)
       image_copy_path = os.path.join(TARGET_DIR,filename)
       shutil.copy2(image_path,image_copy_path)
      
        
print(cnt)
```

```python
# Sampling of images from Kaggle

import random
KAGGLE_FILE_PATH = "COVID/chest_xray/train/NORMAL"
TARGET_NORMAL_DIR = "DataSet/Normal"
```

```python
image_names = os.listdir(KAGGLE_FILE_PATH)
```

```python
image_names
```

```python
random.shuffle(image_names)
```

```python
for i in range(142):
    
    image_name = image_names[i]
    image_path = os.path.join(KAGGLE_FILE_PATH,image_name)
    target_path = os.path.join(TARGET_NORMAL_DIR,image_name)
    shutil.copy2(image_path,target_path)
    print("Copying image ", i)
```
